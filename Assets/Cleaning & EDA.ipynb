## CITY NOT PERFORM
Tujuan: Mengukur performa setiap kota berdasarkan rata-rata total revenue toko di kota tersebut (total revenue kota / total toko di kota itu).

```python
# Grouping data by city, and:
## Calculate Total Revenue.
## Calculate Store Total.

perform1 = mt_cty.groupby('city').agg({
    'cum_revenue': 'sum',
    'store_name': 'nunique'
})

# Added average city revenue data (total city revenue divided by the number of stores).
# Reset index on DataFrame.

perform1['avg_city_revenue'] = round(perform1['cum_revenue'] / perform1['store_name'])
perform1 = perform1.reset_index()

# Sort data by variable "avg_city_revenue" (average city revenue).
# Displays New DataFrame.
perform1 = perform1.sort_values(by=['avg_city_revenue'], ascending=False)
perform1
```

```python
# Determines the size of the canvas to be created.
# Create a Bar Chart using Seaborn.
plt.figure(figsize=(15, 6))
sns.barplot(x='city', y='avg_city_revenue', data=perform1, palette='viridis')

# Set the Bar Chart display.
plt.title('Average total revenue of the city (Total revenue / total stores)')
plt.xlabel('City')
plt.ylabel('Average total Revenue of the city ($)')
plt.xticks(rotation=60)

# Display the Bar Chart.
plt.show()
```

###Insight:

Top 5 cities with highest average total revenue are Ciudad de Mexico, Guadalajara, Toluca, Monterrey, Xalapa, and Hemosillo. These cities seem to have a strong demand for toys, which could indicate higher population densities, higher disposable incomes, or a higher concentration of potential customers interested in buying toys.

Meanwhile cities with the lowest average total revenue are Zacatecas, Tuxtla Gutierrez, Durango, Cuernavaca, and La Paz. These cities might have comparatively lower demand for toys or face other challenges that impact toy store performance, such as competition from other retail sectors or economic factors affecting consumer spending.

```python
# Ubah month_year menjadi tipe datetime
growth_loc_q = monthly_loc.copy()
growth_loc_q['month_year'] = pd.to_datetime(growth_loc_q['month_year'], format='%b-%y')

# Ekstrak tahun dan kuartal
growth_loc_q['year'] = growth_loc_q['month_year'].dt.year
growth_loc_q['quarter'] = growth_loc_q['month_year'].dt.quarter

# Ekstrak pendapatan pada kuartal 1 tahun 2017 dan tahun 2018 untuk lokasi "Airport"
revenue_2017 = growth_loc_q[(growth_loc_q['loc'] == 'Airport') & (growth_loc_q['year'] == 2017) & (growth_loc_q['quarter'] == 1)]['revenue'].values[0]
revenue_2018 = growth_loc_q[(growth_loc_q['loc'] == 'Airport') & (growth_loc_q['year'] == 2018) & (growth_loc_q['quarter'] == 1)]['revenue'].values[0]

# Hitung pertumbuhan pendapatan
revenue_growth_airport = ((revenue_2018 - revenue_2017) / revenue_2017) * 100

print(f"Revenue growth at Airport in Q1 2018 compared to Q1 2017: {revenue_growth_airport:.2f}%")
```

```python
# Function to calculate revenue growth
def calculate_revenue_growth(location, year1, year2, quarter):
    revenue_1 = growth_loc_q[(growth_loc_q['loc'] == location) & (growth_loc_q['year'] == year1) & (growth_loc_q['quarter'] == quarter)]['revenue'].values[0]
    revenue_2 = growth_loc_q[(growth_loc_q['loc'] == location) & (growth_loc_q['year'] == year2) & (growth_loc_q['quarter'] == quarter)]['revenue'].values[0]
    growth = ((revenue_2 - revenue_1) / revenue_1) * 100
    return growth

# Locations to be analyzed
locations = ['Airport', 'Commercial', 'Downtown', 'Residential']

# Years and quarter for comparison
year_1 = 2017
year_2 = 2018
quarter = 1

# Calculate and display revenue growth for each location
for location in locations:
    growth_q1 = calculate_revenue_growth(location, year_1, year_2, quarter)
    print(f"Revenue growth at {location} in Q{quarter} {year_2} compared to Q{quarter} {year_1}: {growth_q1:.2f}%")
```

```python
# Function to calculate total revenue growth for quarters 1, 2, and 3
def calculate_total_revenue_growth(location, year1, year2, quarters):
    total_revenue_1 = growth_loc_q[(growth_loc_q['loc'] == location) & (growth_loc_q['year'] == year1) & (growth_loc_q['quarter'].isin(quarters))]['revenue'].sum()
    total_revenue_2 = growth_loc_q[(growth_loc_q['loc'] == location) & (growth_loc_q['year'] == year2) & (growth_loc_q['quarter'].isin(quarters))]['revenue'].sum()
    growth = ((total_revenue_2 - total_revenue_1) / total_revenue_1) * 100
    return growth

# Locations to be analyzed
locations = ['Airport', 'Commercial', 'Downtown', 'Residential']

# Years and quarters for comparison
year_1 = 2017
year_2 = 2018
quarters = [1, 2, 3]

# Calculate and display total revenue growth for each location
for location in locations:
    growth = calculate_total_revenue_growth(location, year_1, year_2, quarters)
    print(f"Total revenue growth at {location} in Q1-Q3 {year_2} compared to Q1-Q3 {year_1}: {growth:.2f}%")
```

```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Locations to be analyzed
locations = ['Airport', 'Commercial', 'Downtown', 'Residential']

# Years and quarters for comparison
year_1 = 2017
year_2 = 2018
quarters = [1, 2, 3]

# Calculate growth and create subplots for each location
for idx, location in enumerate(locations):
    growth_quarters = []
    for quarter in quarters:
        growth = calculate_revenue_growth(location, year_1, year_2, quarter)
        growth_quarters.append(growth)
    total_growth = calculate_total_revenue_growth(location, year_1, year_2, quarters)

    row = idx // 2
    col = idx % 2

    ax = axes[row, col]
    ax.plot(quarters, growth_quarters, marker='o', label='Quarterly Growth 2017 compare with 2018')
    ax.plot(4, total_growth, marker='s', color='red', label='Total Growth 2017 compare with 2018')

    ax.set_title(f'Compare Growth 2017 with 2018 in {location}')
    ax.set_xlabel('Quarter')
    ax.set_ylabel('Growth (%)')
    ax.set_xticks(quarters + [4])
    ax.set_xticklabels(['Q1', 'Q2', 'Q3', 'Total'])
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
from tabulate import tabulate

# Create a dictionary to store the results
results = []

# Locations to be analyzed
locations = ['Airport', 'Commercial', 'Downtown', 'Residential']

# Years and quarters for comparison
year_1 = 2017
year_2 = 2018
quarters = [1, 2, 3]

# Calculate growth for each location
for location in locations:
    growth_quarters = [calculate_revenue_growth(location, year_1, year_2, quarter) for quarter in quarters]
    total_growth = calculate_total_revenue_growth(location, year_1, year_2, quarters)

    results.append({
        'Location': location,
        'Q1 Growth': growth_quarters[0],
        'Q2 Growth': growth_quarters[1],
        'Q3 Growth': growth_quarters[2],
        'Total Growth': total_growth
    })

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Print the results as a formatted table
table = tabulate(results_df, headers='keys', tablefmt='grid')
print(table)
```
